
# **What is Data visualization?**

1. Human brain can process data easily when it is in pictorial or graphical format.
2. The data in graphical format allows to identify new trends and patterns easily.

# **1. Line chart**

Example: Suppose you want to check sales of a departmental store in last 5 years.

```python
import matplotlib.pyplot as plt
year = [2016,2017,2018,2019,2020]
sales = [20,40,30,50,10]
#plt.plot(year,sales)
#plt.plot(year,sales,marker='*')
plt.plot(year,sales,marker='*',color='g',ls=':')
plt.show()
```

#**Format Strings**
You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

**marker|line|color**

```python
year = [2016,2017,2018,2019,2020]
sales = [20,40,30,50,10]
#plt.plot(year,sales,marker='*',color='g',ls=':')
plt.plot(year,sales,'*:g')  #marker|line|color
plt.show()
```

**title and labels:**

You can add a title to each plot with the **title()** function:

you can use the **xlabel()** and **ylabel()** functions to set a label for the x- and y-axis.

```python
import matplotlib.pyplot as plt
year = [2016,2017,2018,2019,2020]
sales = [20,40,30,50,10]
plt.plot(year,sales)
plt.xlabel('year')
plt.ylabel('sales')
plt.title('sales per year',loc='center') #by default: centre
# loc = 'left' || loc = 'right' || loc = 'centre' 

plt.show()
```

#**Display Multiple Plots**
image.png

With the **subplots()** function you can draw multiple plots in one figure:

```python
import matplotlib.pyplot as plt
#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

plt.subplot(1, 2, 1)
#the figure has 2 row, 1 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x1 = [3, 4, 5, 6]
y1 = [10, 20, 30, 40]

plt.subplot(1, 2, 2)
#the figure has 2 row, 1 columns, and this plot is the second plot.
plt.plot(x1,y1)

plt.show()
```

You can draw as many plots you like on one figure, just descibe the number of rows, columns, and the index of the plot.



# **2. Scatter Plots**

To observe and show `relationships` between two numeric variables. 

The dots in a scatter plot represent the values of individual data points.

Capture.JPG

```python
year = [2016,2017,2018,2019,2020]
sales = [20,40,50,60,70]


plt.scatter(year,sales)
#plt.scatter(year,sales,color='green',s=100, alpha=1)
#color = "red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"
#s: size
#alpha: can adjust the transparency of the dots with the alpha argument.
plt.xlabel('year')
plt.ylabel('sales')
plt.title('sales per year')


plt.show()
```

# **3. Bar Plot**


A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.



 Example:If we want to show the number of students enrolled for various courses offered at an institute.

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
#plt.bar(langs,students)
#plt.barh(langs,students)  #horizontal
#color=
#width=0.1,..

plt.bar(langs,students,color='red',width=.1)

plt.show()
```

# **4. Histogram**
A histogram is an accurate representation of the distribution of **numerical data**.

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

**To construct a histogram, follow these steps −**

Bin the range of values.

Divide the entire range of values into a series of intervals.

Count how many values fall into each interval.

```python
# marks obtained by students in a class
from matplotlib import pyplot as plt
import numpy as np

marks = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(marks,bins = [0,20,40,60,80,100])
plt.show()
```

# **5. Pie Chart**
A Pie Chart can only display one series of data.

The data points in a pie chart are shown as a percentage of the whole pie.



The proportionate percentage is displayed inside the respective wedge/part with the help of **autopct** parameter

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

#plt.pie(students)
plt.pie(students,labels=langs,autopct='%.2f%%')  ##decimal value upto one place
plt.show()
```
